### Adjust the data format to float

```python
# Changing the data type of discounted price and actual price
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')

# Changing Datatype and values in Discount Percentage
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')

df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
# Finding unusual string in rating column
df['rating'].value_counts()
```

```python
# Check the strange row
df.query('rating == "|"')

# Changing Rating Columns Data Type

df['rating'] = df['rating'].str.replace('|', '3.9').astype('float64')

# I got this product rating on Amazon by searching the provided product_id on their official website (amazon.in)¶
# The rating is 3.9. So, I am going to give the item rating a 3.9 as well.¶
```

```python
# Changing 'rating_count' Column Data Type

df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

## 3.Data Cleaning


### Missing Values

```python
# 1.Check missing values
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_data = pd.concat([missing_values, missing_percentage], axis=1)
missing_data.columns = ['缺失值数量', '缺失值百分比']
print(missing_data)
```

### review content analysis


```python
from textblob import TextBlob
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
# simplified text preprocessing function
from nltk.corpus import stopwords
import re

def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', str(text))
    tokens = text.lower().split()
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [token for token in tokens if token not in stop_words] # Stop words removed
    cleaned_text = " ".join(filtered_tokens)
    if not cleaned_text:  # Check if the string is empty
        return ""  # Or some other placeholder like "empty_review"
    return cleaned_text
```

```python
# apply the preprocessing function
df['clean_review'] = df['review_content'].apply(preprocess_text)
```

```python
df['clean_review'].head()
```

```python
# sentiment analysis
# simplified text preprocessing function
def simple_preprocess_text(text):
    # remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', str(text))
    # simple tokenization
    return text.lower().split()

def get_sentiment(text):
    analysis = TextBlob(' '.join(text))
    return analysis.sentiment.polarity
```

```python
df['simple_clean_review'] = df['review_content'].apply(simple_preprocess_text)
df['sentiment'] = df['simple_clean_review'].apply(get_sentiment)
```

```python
# keyword extraction(using TF-IDF)
df['clean_review'] = df['review_content'].apply(preprocess_text)

# IMPORTANT: Remove rows with empty reviews *before* TF-IDF
df = df[df['clean_review'] != ""]

tfidf = TfidfVectorizer(max_features=50, stop_words='english')  # Or use a custom stop word list
tfidf_matrix = tfidf.fit_transform(df['clean_review'])  # Use clean_review
feature_names = tfidf.get_feature_names_out()

print(feature_names) # Check the keywords
```

```python
# ========== LDA model ==========
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# use the already cleaned review data
processed_reviews = [' '.join(review) for review in df['simple_clean_review'] if len(review) > 0]

# document term matrix
cv = CountVectorizer(
    max_features=1000, 
    stop_words='english',
    min_df=2  # at least appear in 2 documents
)
doc_term_matrix = cv.fit_transform(processed_reviews)

# LDA model
lda = LatentDirichletAllocation(
    n_components=5, 
    random_state=42,
    max_iter=10
)
lda_output = lda.fit_transform(doc_term_matrix)

# show the topic words
feature_names = cv.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
    top_words = [feature_names[i] for i in topic.argsort()[:-10:-1]]
    print(f"topic {topic_idx + 1}:")
    print(", ".join(top_words))
    print()
```
